```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# Load the CSV file
def load_csv(file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    return df

# Combine Description and Patient columns
def combine_columns(df):
    df["combined"] = (
        "Description: " + df["Description"].str.strip() +
        "; Patient: " + df["Patient"].str.strip()
    )
    return df

# Generate embeddings for the combined column
def embed_text(df, text_col, model_name="all-MiniLM-L6-v2"):
    # Load the Hugging Face SentenceTransformer model
    model = SentenceTransformer(model_name)

    # Generate embeddings
    df["embeddings"] = df[text_col].apply(lambda x: model.encode(x))
    return df

# Main function to process the file and generate embeddings
def main():
    # Path to your CSV file
    file_path = "C:\\HAP774\\New folder (2)\\MedLlama_Chatbot\\cleaned_data.csv"

    # Step 1: Load the data
    df = load_csv(file_path)

    # Step 2: Combine the Description and Patient columns
    df = combine_columns(df)

    # Step 3: Generate embeddings for the combined column
    df = embed_text(df, text_col="combined")

    # Save the DataFrame with embeddings
    output_path = "C:\\HAP774\\New folder (2)\\MedLlama_Chatbot\\cleaned_data_with_embeddings.pkl"
    df.to_pickle(output_path)
    print(f"Embeddings saved to {output_path}")

# Run the script
if __name__ == "__main__":
    main()
```
